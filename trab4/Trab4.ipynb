# Trabalho 4 - Operações Geométricas

```python
%matplotlib inline

import cv2
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("imagem.jpg", 0)
maxWidth, maxHeight = img.shape

init_points = np.matrix([[550,  110],
                         [40,   1050],
                         [1260, 1300],
                         [1360, 250]], dtype='float32')

transf_points = np.matrix([[0,          0],
                           [0,          maxHeight-1],
                           [maxWidth-1, maxHeight-1],
                           [maxWidth-1, 0]], dtype='float32')

T = cv2.getPerspectiveTransform(init_points, transf_points)
transf = cv2.warpPerspective(img, T, (maxWidth, maxHeight))

plt.subplot(121)
plt.imshow(img, cmap=plt.get_cmap('gray'), vmin=0, vmax=255)
plt.title('Original')
plt.subplot(122)
plt.imshow(transf, cmap=plt.get_cmap('gray'), vmin=0, vmax=255)
plt.title('Transformada')
```
