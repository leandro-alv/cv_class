# Top Hat

```python
%matplotlib inline

import cv2
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('imagem.jpg',0)

kernel = np.ones((5,5),np.uint8)
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)

plt.subplot(121)
plt.imshow(img, cmap=plt.get_cmap('gray'), vmin=0, vmax=255)
plt.title('Original')
plt.subplot(122)
plt.imshow(tophat, cmap=plt.get_cmap('gray'), vmin=0, vmax=255)
plt.title('Top Hat')
```
